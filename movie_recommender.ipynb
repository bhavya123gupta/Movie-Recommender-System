```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# Covert it into like this:-
# ['Action','Adventure','Fantasy','SciFi']
# Applying pre-processing(Step 1) 
# Function to convert JSON like string to list of genre names
import ast 
```

```python
# function to convert JSON-like string to list of names
def convert(obj):
    L = []
    obj = ast.literal_eval(obj) # convert JSON-like string to actual list of dictionaries 
    for i in obj:
        L.append(i['name'])
    return L 
# Applying the convert function to the genres column 
movies['genres'] = movies['genres'].apply(convert)
# Example of accessing genres of the first row
print(movies.iloc[0]['genres'])
# Printing all the rows 
print(movies['genres'])
movies['keywords'] = movies['keywords'].apply(convert)
print(movies['keywords'])
```

```python
# function to get top 3 cast members
def convert3(obj):
    L = []
    obj = ast.literal_eval(obj) # Convert JSON-like string to actual list of dictionaries
    counter = 0
    for i in obj:
        if(counter!=3):
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
movies['cast'] = movies['cast'].apply(convert3)
print(movies['cast'])

 
```

```python
# in 'crew' we have to extract job = director and that director's name from the same
# dictionary(list) among a lot of dictionaries(lists).
# Ek dictionary mei ek job description ke insaan ke baare mei details hogi
# hume director ke job wale insaan ki dictionary(list) mei se uska name extract karna hai

def fetch_director(obj):
    L =[]
    obj = ast.literal_eval(obj)
    for i in obj:
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
movies['crew'] = movies['crew'].apply(fetch_director)
print(movies['crew'])
```

```python
def incoporate_moods(row):
    moods_tags = []
    genres = row['genres']
    
    if 'Comedy' in genres or 'Adventure' in genres or 'Animation' in genres:
        moods_tags.append('happy')
    if 'Drama' in genres or 'Romance' in genres or 'War' in genres:
        moods_tags.append('sad')
    if  'Drama' in genres or 'Romance' in genres or 'Music' in genres or 'War' in genres:
        moods_tags.append('heart touching')
    if 'History' in genres or 'Animation' in genres or 'Music' in genres or 'Family' in genres:
        moods_tags.append('nostalgic')
    if 'Comedy' in genres:
        moods_tags.append('funny')
    if 'Adventure' in genres:
        moods_tags.append('adventurous')
    if 'Romance' in genres:
        moods_tags.append('romantic')
    if 'Mystery' in genres:
        moods_tags.append('mysterious')
    if 'Horror' in genres or 'Mystery' in genres:
        moods_tags.append('scared')
    if 'Action' in genres or 'Thriller' in genres:
        moods_tags.append('excited')
    if 'Thriller' in genres: 
        moods_tags.append('thrilling')
    if 'Fantasy' in genres or 'Family' in genres:
        moods_tags.append('relaxed')
    return moods_tags
    
movies['moods'] = movies.apply(incoporate_moods, axis=1)
movies['moods'] = movies['moods'].apply(lambda x: " ".join(x))
movies.head()
```

```python
def get_recommendations(mood):
    
    # check if 'moods' column exists
    if 'moods' not in movies.columns:
        print("Error: 'moods' column not found in the movies Dataframe.")
        return[]
    else:
        print("Moods column is there")
    
    # Define a mapping between moods and genres
    mood_to_genre = {
        'happy': ['Comedy', 'Family'],
        'sad': ['Drama', 'Romance'],
        'scared': ['Horror'],
        'excited': ['Action', 'Adventure'],
        'funny': ['Comedy'],
        'relaxed': ['Drama', 'Romance'],
        'romantic': ['Romance'],
        'adventurous': ['Adventure'],
        'thrilling': ['Thriller', 'Action'],
        'mysterious': ['Mystery', 'Thriller'],
        'nostalgic': ['Drama', 'Family'],
        'heart touching': ['Drama', 'Romance']
    }

    # Get genres associated with the selected mood
    genres_for_mood = mood_to_genre.get(mood.lower(), [])
    
    filtered_movies = movies[movies['moods'].apply(lambda x: mood in x)]
    if filtered_movies.empty:
        return[]
    # Inspecting the tags after filtering
    #print("Filtered tags columns: ")
    #print(filtered_movies['tags'])
    
    # Check if the filtered tags are empty or contain only stop words
    if filtered_movies['tags'].str.strip().eq('').all():
        print(f"No tags found for the movies in the {mood} mood.")
        return []
        
    # Vectorize the tags of filtered movies
    cv = CountVectorizer(max_features=5000, stop_words='english')
    try:
         filtered_vectorized_tags = cv.fit_transform(filtered_movies['tags'])
    except ValueError as e:
        print(f"Error during vectorization: {e}")
        return []
        
    # check if vocabulary is empty after vectorization 
    if len(cv.get_feature_names_out()) == 0:
        print(f"Empty vocabulary after filtering the tags for the {mood} mood.")
        return []
        
    # Calculate cosine similarity for filtered movies
    filtered_cosine_sim = cosine_similarity(filtered_vectorized_tags)

    # Get the top 10 movie recommedations
    similar_movie_indices = filtered_cosine_sim.mean(axis=0).argsort()[::-1][:10]
    #Return movie titles based on the similar movie indices
    recommended_movies = filtered_movies.iloc[similar_movies_indices]['title'].tolist()

    return recommended_movies
```

```python
# Weekly top 10 movies 
def get_weekly_top_10():
    top_10_movies = movies.sort_values(by='popularity', ascending=False).head(10)
    top_10_titles = top_10_movies['title'].values
    return top_10_titles
print( get_weekly_top_10())    
```

```python
# User Wishlists Management 
class Wishlist:
    def __init__(self):
        self.wishlist = []
        
    def add_to_wishlist(self, movie_title):
        user_response = input(f"Do you want to add '{movie_title}' to your wishlist? (yes/no): ").strip().lower()
        if user_response == 'no':
            print("Ok, no problem.")
        elif user_response == 'yes':    
            self.wishlist.append(movie_title)
            print(f'"{movie_title}" has been added to your wishlist.')
            self.show_wishlist_prompt()
            
    def show_wishlist_prompt(self):
        view_response =input("here is your watchlist...Do you want to view it? (yes/no): ").strip().lower()
        if view_response == 'yes':
            self.show_wishlist()
    def show_wishlist(self):
        if not self.wishlist:
            print("Your watchlist is empty.")
        else:
            print("Your watchlist: ")
            for movie in self.wishlist:
                print(movie)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to ask user for mood and recommend movies 
def recommend_movies_for_user():
    print("Please select your mood:")
    for i, mood in enumerate(moods, 1):
        print(f"{i}. {mood}")
    selected_mood_index = int(input("Enter the number corresponding to your mood: "))
    selected_mood = moods[selected_mood_index -1]

    recommendations = get_recommendations(selected_mood)
    if recommendations:
        print(f"\nMovies recommended for your {selected_mood} mood: ")
        for movie in recommendations:
            print(movie)
    else:
        print(f"No recommendation found for the {selected_mood} mood.")
    # Weekly top 10 movies 
    weekly_top_10 = get_weekly_top_10()
    print("\nWeekly Top 10 Movies: ")
    for movie in weekly_top_10:
        print(movie)   
    # Wishlist Functionality 
    wishlist = Wishlist()
    while True:
        print("\nDo you want to add any movie in your wishlist?? (yes/no)")
        add_to_wishlist = input().lower()
        if add_to_wishlist == 'yes':
            print("Enter the movie title to add to your wishlist: ")
            movie_title = input()
            wishlist.add_to_wishlist(movie_title)
            print("Your movie has been added to wishlist... thanks for adding it...making your experience better!!!")
        elif add_to_wishlist == 'no':
            print(" Ok, no problem")
            break
        else:
            print("Please enter a valid response (yes/no).")
    print("\nYour Wishlist:")
    wishlist.view_wishlist()
# Run the recommender system
recommend_movies_for_user()
```
